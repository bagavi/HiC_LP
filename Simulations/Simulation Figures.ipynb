```python
n = 1000
it_limit = np.sqrt(4*np.log(n))
mg_limit = np.sqrt(6*np.log(n))
sg_limit = np.sqrt(8*np.log(n))
filenameBP = 'data/BP_'+str(n)+'.csv'   
filenameLP = 'data/LP_'+str(n)+'.csv'
filenameMG = 'data/MG_'+str(n)+'.csv'

f = open(filenameBP, 'rb')
readerBP = csv.reader(f)
resultsBP  = []
for i, row in enumerate(readerBP):
    if len(row)!=n*10+2:
        print len(row), i, row[:2]
        continue
    resultsBP  += [np.array(row, dtype='float')]
resultsBP  = np.array(resultsBP)


f = open(filenameLP, 'rb')
readerLP = csv.reader(f)
resultsLP  = []
for row in readerLP:
    resultsLP  += [np.array(row[:-1], dtype='float')]
resultsLP  = np.array(resultsLP)


f = open(filenameMG, 'rb')
readerMG = csv.reader(f)
resultsMG  = []
for row in readerMG:
    resultsMG  += [np.array(row[:-1], dtype='float')]
resultsMG  = np.array(resultsMG)


if not np.mean(np.unique(resultsBP [:,1]) == np.unique(resultsLP [:,1])):
    print "PLS CHECK!"
mu_array = np.unique(resultsBP [:,1])
mu_arrayMG = np.unique(resultsMG [:,1])
mu_max = mu_array.max()
mu_min = mu_array.min()
print "Sample sizes", resultsBP.shape[0], resultsLP.shape[0], resultsMG.shape[0]
```

```python
mu_array= mu_array[:-12]
# mu_array
```

```python
LP_dict = dict()
mu_count = dict()
for row in resultsLP:
    mu = row[1]
    try:
        LP_dict[mu] += [row[2]/6.+row[3]/1.5]
    except:
        LP_dict[mu] = [row[2]/6.+row[3]/1.5]
    try:
        mu_count[mu] += 1
    except:
        mu_count[mu] =0        
for mu in mu_array:
    LP_dict[mu] = np.array(LP_dict[mu])
plt.figure(figsize=(10,4))
plt.plot(mu_count.keys(), mu_count.values(), 'o')
plt.show()
```

```python
BP_dict = dict()
mu_count = dict()
for row in resultsBP:
    mu = row[1]
    try:
        BP_dict[mu] += [row[3:]]
    except:
        BP_dict[mu] = [row[3:]]
    try:
        mu_count[mu] += 1
    except:
        mu_count[mu] =0        
for mu in mu_array:
    BP_dict[mu] = np.array(BP_dict[mu])
plt.figure(figsize=(10,4))
plt.plot(mu_count.keys(), mu_count.values(), 'o')
plt.show()
```

```python
MG_dict = dict()
mu_count = dict()
for row in resultsMG:
    mu = row[1]
    try:
        MG_dict[mu] += [row[3]]
    except:
        MG_dict[mu] = [row[3]]
    try:
        mu_count[mu] += 1
    except:
        mu_count[mu] =0        
for mu in mu_arrayMG:
    MG_dict[mu] = np.array(MG_dict[mu])
plt.figure(figsize=(10,4))
plt.plot(mu_count.keys(), mu_count.values(), 'o')
plt.show()
```
