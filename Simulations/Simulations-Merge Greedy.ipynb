```python
def Solve_MG(data):
    n,_ = data.shape
    answer = np.zeros_like(data)
    # Greedy but cautious
    count_arr = np.zeros(n)
    order = np.dstack(np.unravel_index(np.argsort(data.ravel()), data.shape))[0][::-1]
    assigned = 0
    for i, row in enumerate(order):
        if row[0] > row[1]: #each entry is present twice 
            continue
        if count_arr[row[0]] <2 and  count_arr[row[1]]<2:
            answer[row[0]][row[1]]=1
            answer[row[1]][row[0]]=1
            count_arr[row[0]]    +=1
            count_arr[row[1]]    +=1
            assigned +=1
        if assigned >n*2+1:
            break
    return answer

def MG(data, file):
    n = data.shape[0]
    answer      = Solve_MG(data)
#     print answer
    correct_answer = algorithms.correct_data_answer(data,k=1)
           
    error = (np.where(correct_answer-answer!=0)[0].shape[0]+0.0)/(2*n)
    fp = (np.where(correct_answer-answer==-1)[0].shape[0]+0.0)/(2*n)
    point5 = (np.where(answer==0.5)[0].shape[0]+0.0)/(2*n)
    if file != None:
        fd = open(file,'a')
        fd.write(', '.join([str(n), str(mu), str(error), str(fp), str(point5),"\n"]))
        fd.close()
```
