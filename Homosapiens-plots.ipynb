```python
import visualize, csv
import numpy as np
from collections import Counter 
%matplotlib inline
folder = 'real_dataset_experiments/HomoSapiens/error_rates/'
```

```python
chromosome = str(1)
filepath     =  folder+"100000_LP_greedy_chr"+chromosome+".csv"

f = open(filepath, 'rb')
reader = csv.reader(f)
results = []
for row in reader:
    results += [row[:-1]]
results = np.array(results)
```

```python
c = Counter(results[:,-1])
mu_array = np.zeros((20,2))
for j, i in enumerate(c):
    mu_array[j][0] = float(i[1:])
    mu_array[j][1] = float(c[i])
mu_array = mu_array[np.argsort(mu_array[:,0])]
a = np.cumsum(mu_array[:,1])
for i in range(2,19):
    zz =  np.array(results[a[i]:a[i+1]][:,0], dtype='float')/2.
    print int(mu_array[i][0]*100)/100., "\t",int(zz.mean()*100)/100., int(zz.std()*1000)/1000.
```
